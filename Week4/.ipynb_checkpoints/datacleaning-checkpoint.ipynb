```python
# Identify any null values in the DataFrame using the isnull function.

ppl_reordered_csv.isnull()


```

```python
# Determine the number of nulls in the DataFrame using the isnull function in conjunction with the sum function.

ppl_reordered_csv.isnull().sum()
```

```python
# Determine the percentage of nulls for each column by chaining the isnull and sum functions then dividing by the len of the DataFrame. Multiply by 100 to produce a percentage rather than a decimal number.

pct_nulls = ppl_reordered_csv.isnull().sum() / len(ppl_reordered_csv) * 100
pct_nulls
```
