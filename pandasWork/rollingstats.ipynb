```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
import seaborn as sb
from pathlib import Path

%matplotlib inline

csv_path = ("MSFT.csv")

csv_path = pd.read_csv(csv_path, index_col="date", infer_datetime_format=True, parse_dates=True)
csv_path.head()
```

```python
#Drop the extra columns: 'volume', 'open', 'high', 'low'.

csv_path.drop(["volume","open","high","low"], axis=1, inplace=True)
csv_path.head()
```

```python
#Use the rolling, mean, and plot functions to calculate a 7 day rolling window of average MSFT closing prices.
#df.close.rolling(3).mean()
roll_mean = csv_path.close.rolling(7).mean()
roll_mean.plot()
```

```python
# Plot a rolling 7-day std dev of MSFT's closing prices
roll_stddev = csv_path.rolling(window=7).std()
roll_stddev.plot()
```

```python
## 8. Rolling 30-Day Mean of MSFT Closing Prices
roll_mean_30 = csv_path.rolling(window=30).mean()
roll_mean_30.plot()
```

```python
#### 9. Rolling 30-Day Standard Deviation of MSFT Closing Prices
roll_std_30 = csv_path.rolling(window=30).std()
roll_std_30.plot()
```

```python
## 10. Rolling 180-Day Mean of MSFT Closing Prices
roll_mean_180 = csv_path.rolling(window=180).mean()
roll_mean_180.plot()
```

```python
## 11. Rolling 180-Day Standard Deviation of MSFT Closing Prices

roll_std_180 = csv_path.rolling(window=180).std()
roll_std_180.plot()
```

```python
## 12. Overlay Daily Closing Price and 180-Day Rolling Mean of MSFT
# Set figure of the daily closing prices of MSFT

ax = csv_path.plot()

# Plot 180-Day Rolling Mean on the same figure
csv_path.rolling(window=180).mean().plot(ax=ax)

# Set the legend of the figure
ax.legend(["MSFT", "MSFT 180 Day Mean"]);
```
