```python
# Initial imports
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline

# Take sample of data
csv_path = Path("whale_returns.csv")
csv_data = pd.read_csv(csv_path, index_col="Date")
csv_data.head()
```

```python
csv_data.head()
csv_data.isnull()
```

```python
#count nulls
csv_data.isnull()
```

```python
# describe
csv_data.describe(include='all')
```

```python
# drop nulls
csv_data = csv_data.dropna()
csv_data.head()
```

```python
csv_data.isnull()
```

```python
#reading algorithmic returns - index already set to the date column
#working on Soros Fund Management LLC

csv_data = csv_data.drop(columns=["TIGER GLOBAL MANAGEMENT LLC","BERKSHIRE HATHAWAY INC"])


```

```python
csv_data = csv_data.drop(columns=["SOROS FUND MANAGEMENT LLC"])
csv_data.head()
```

```python
#read algorithmic returns - daily returns on Soros

daily_returns = (csv_data - csv_data.shift(1)) / csv_data.shift(1)
daily_returns.head()
```

```python
daily_returns.plot()
```

```python
#cumulative returns
cumulative_return = (1 + daily_returns).cumprod()
```
