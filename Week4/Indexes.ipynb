```python
#Slice and Dice Pandas DataFrames via Indexes
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

cleaned_people_csv = Path("cleaned_people_data.csv")
cleaned_people_csv = pd.read_csv(cleaned_people_csv,index_col=0)
cleaned_people_csv.head()
```

```python
#cleaned_people_csv.iloc([1:1])
cleaned_people_csv.iloc[[1,2]]
```

```python
cleaned_people_csv.iloc[[1,2,3,4,5,6,7,10]]
```

```python
#select first 10 rows using iloc
first_n_rows = cleaned_people_csv.iloc[:10]
first_n_rows
```

```python
# Select the last 10 rows of the DataFrame using iloc.

last_n_rows = cleaned_people_csv.iloc[-10:]
last_n_rows
```

```python
#Select the second column using iloc.

second_column = cleaned_people_csv.iloc[:,:2]
second_column
```

```python
# Select the last column using iloc

last_n_column  = cleaned_people_csv.iloc[: , -2:]
last_n_column
```

```python
# Select the first three columns using iloc.
first_3_column  = cleaned_people_csv.iloc[: ,1-3:]
first_3_column
```

```python
# Select the 1st, 3rd, 5th, and 7th rows of the 2nd, 4th, and 6th columns using iloc.
#df. iloc[:, [0,1,3]]

first_multi_column  = cleaned_people_csv.iloc[[1,3,5,7],[4,6]]
first_multi_column
```

```python
#Select the first 5 rows of the 3rd, 4th, and 5th columns using iloc.

first_5  = cleaned_people_csv.iloc[:5],[3,4,5]
first_5

```

```python
# Modify the first_name column value of the first row using get_loc and iloc

first_val = cleaned_people_csv.iloc[:1]
first_val
```
